```python
%%time

paths = [str(x) for x in Path(dataset_path).glob("*.txt")]
print("Datadir:", dataset_path)
print("paths:", paths)
```

```python
# this need token: [SEP], [CLS], unk, mask
tokenizer_bert = BertWordPieceTokenizer(strip_accents=False, lowercase=False, \
                                        handle_chinese_chars=False, clean_text=False)
tokenizer_bert.train(files=paths, vocab_size=104000, min_frequency=2,
                show_progress=True,
    special_tokens=["[PAD]", "[UNK]", "[CLS]", "[SEP]", "[MASK]"],
    limit_alphabet=1000,
    wordpieces_prefix="##")
```
